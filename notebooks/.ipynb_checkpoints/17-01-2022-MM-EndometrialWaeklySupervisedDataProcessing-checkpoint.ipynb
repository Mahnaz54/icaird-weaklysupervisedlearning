```python
import pandas as pd
import os
import sys
import glob
from xlrd import * 
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
#READ CEXCEL FILES AND CONVERT THEM TO CSV FILES AND COMBINE THEM TO ONE FILE
root = '/home/mahnaz/Desktop/iCAIRD_new_slide_labels'
excel_files_dir  = os.path.join(root + '/excel_files')
csv_files_dir = os.path.join(root + '/csv_files')

filenames = os.listdir(excel_files_dir)
for filename in filenames:
    #print('processing {}'.format(filename))
    if '.xls' in filename:
        read_file = pd.read_excel( excel_files_dir + '/'+ filename)
        read_file.to_csv (csv_files_dir +'/' +filename.split('.')[0] + '.csv', index = None, header=True, encoding = 'utf-8-sig')
        
all_filenames = os.listdir(csv_files_dir)
```

```python
#combine all files in the list
df_all = pd.concat([pd.read_csv(csv_files_dir+ '/'+ f) for f in all_filenames ])
#print('Total umber of slides: {}'.format(len(df_all)))
total_endometrial = df_all[df_all.SampleType =='EN']
```

```python
excluded_endo = total_endometrial[total_endometrial.ExcludedFromAnnotation==1]
excluded_endo.to_csv(root+ '/' + 'weaklysupervised_algo' + '/' +'excluded_endo_files.csv', index= False)
```

```python
endo = total_endometrial[total_endometrial.ExcludedFromAnnotation==0]
endo = endo.sort_values(by ='ImageName')
endo['Cat'] = endo['Cat'].str.lower()
endo['SubCat'] = endo['SubCat'].str.lower()


endo['SubCat'] = endo['SubCat'].str.replace('/', '_')
endo['Cat'] = endo['Cat'].str.replace('/','_')
endo['Cat'] =endo['Cat'].str.replace('innactive_atrophic', 'other_benign')

endo['SubCat'] = endo['SubCat'].str.replace('hyperplasia with atypia', 'hyperplaysia')
endo['SubCat'] = endo['SubCat'].str.replace('innactive_atrophic', 'innactive')
endo
```

```python
print('Out of  {} slides , {} are endometiral'.format(len(df_all), len(total_endometrial)))
print('{} endo files are excluded from total {}'.format(len(excluded_endo), len(df_all)))
print('{} endo files are to be processed'.format(len(endo)))
```

```python
print('Total Endo Category Counts:\n', endo['Cat'].value_counts())
```

```python
print('Total Endo Sub Category Counts:\n', endo['SubCat'].value_counts())
```

```python
endo['train/valid/test']= endo['AnnotationBatch'].apply(lambda x: 'train' if 36 <= x <=76 else 'test')
endo
```

```python
print('Total Endo Counts:\n', endo['train/valid/test'].value_counts())
```

```python
endo_clean = endo[['AnnotationBatch','SampleID', 'ImageName', 'Cat', 'train/valid/test']]
endo_clean= endo_clean.rename(columns={'SampleID':'case_id', 'ImageName':'slide_id', 'Cat':'label'})
endo_clean = endo_clean[(endo_clean['case_id'] != 'IC-EN-000684') & (endo_clean['case_id'] != 'IC-EN-02700') & (endo_clean['case_id'] !='IC-EN-03004')]
```

```python
endo_clean.to_csv( root+ '/' + 'weaklysupervised_algo' + '/' + 'endo_clean.csv', index= False)
```

```python
endo_train = endo_clean[endo_clean['train/valid/test'] == 'train']
endo_test = endo_clean[endo_clean['train/valid/test'] == 'test']
from sklearn.model_selection import train_test_split
train, endo_valid = train_test_split(endo_train, test_size=0.25, random_state=0, stratify=endo_train[['label']])
endo_valid['train/valid/test'] = 'valid'

print('train:{}\t valid:{} \t test:{}\n\n'.format(len(train), len(endo_valid), len(endo_test)))

print('Train cat count:\n', train['label'].value_counts())
print('\n Valid cat count:\n', endo_valid['label'].value_counts())
print('\n Test cat count:\n', endo_test['label'].value_counts())
train = train['slide_id']
valid = endo_valid['slide_id']
test = endo_test['slide_id']
endo_split= pd.concat([train.reset_index(drop=True),valid.reset_index(drop=True),test.reset_index(drop=True)], axis = 1, ignore_index=True)
endo_split = endo_split.rename(columns={endo_split.columns[0]: 'train',endo_split.columns[1]: 'val',endo_split.columns[2]: 'test' })
endo_split
```

```python
endo_split.to_csv(root+ '/' + 'weaklysupervised_algo' + '/' + 'split_0.csv', index= False)
```
