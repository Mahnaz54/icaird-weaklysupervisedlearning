```python
import pandas as pd
import os
import sys
import glob
from xlrd import * 
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
#READ CEXCEL FILES AND CONVERT THEM TO CSV FILES AND COMBINE THEM TO ONE FILE
root = '/home/mahnaz/Desktop/iCAIRD_new_slide_labels'
excel_files_dir  = os.path.join(root + '/excel_files')
csv_files_dir = os.path.join(root + '/csv_files')

filenames = os.listdir(excel_files_dir)
for filename in filenames:
    #print('processing {}'.format(filename))
    if '.xls' in filename:
        read_file = pd.read_excel( excel_files_dir + '/'+ filename)
        read_file.to_csv (csv_files_dir +'/' +filename.split('.')[0] + '.csv', index = None, header=True, encoding = 'utf-8-sig')
        
all_filenames = os.listdir(csv_files_dir)
```

```python
#combine all files in the list
df_all = pd.concat([pd.read_csv(csv_files_dir+ '/'+ f) for f in all_filenames ])
#print('Total umber of slides: {}'.format(len(df_all)))

#sperate endometrial and cervical files
total_cervical = df_all[df_all.SampleType =='CX']
total_endometrial = df_all[df_all.SampleType =='EN']


#filter out the excluded files
excluded_cervical = total_cervical[total_cervical.ExcludedFromAnnotation==1]
excluded_cervical.to_csv( root+ '/' + 'supervised_algo' + '/' + 'excluded_cervical_files.csv', index= False)


excluded_endo = total_endometrial[total_endometrial.ExcludedFromAnnotation==1]
excluded_endo.to_csv(root+ '/' + 'supervised_algo' + '/' +'excluded_endo_files.csv', index= False)

endo = total_endometrial[total_endometrial.ExcludedFromAnnotation==0]
cervical = total_cervical[total_cervical.ExcludedFromAnnotation==0]

cervical = cervical.sort_values(by ='ImageName')
endo = endo.sort_values(by ='ImageName')

cervical['Cat'] = cervical['Cat'].str.lower()
cervical['SubCat'] = cervical['SubCat'].str.lower()

cervical['SubCat'] = cervical['SubCat'].str.replace('/', '_')
cervical['Cat'] = cervical['Cat'].str.replace('/','_')

cervical['Cat'] = cervical['Cat'].str.replace(' ','_')
cervical['SubCat'] = cervical['SubCat'].str.replace('cin 1', 'cin1')
cervical['SubCat'] = cervical['SubCat'].str.replace('cin 2', 'cin2')
cervical['SubCat'] = cervical['SubCat'].str.replace('cin 3', 'cin3')
cervical['SubCat'] = cervical['SubCat'].str.replace(' ', '_')

cervical = cervical[cervical.SubCat != 'insufficient']

print('Out of  {} slides , {} are cervical and {} are endometrial'.format(len(df_all), len(total_cervical), len(total_endometrial))) 

print('{} cervical files are excluded'.format(len(excluded_cervical)))
print('{}  endometrial files are excluded'.format(len(excluded_endo)))

print('Total number of cervical  slides after exclusion = {}'.format(len(cervical)))
print('Total number of endometrial  slides after exclusion = {}'.format(len(endo)))

cervical.head()
```

```python
cervical = cervical.rename(columns={'ImageName':'Filename','SampleID':'BiobankID', 'Cat':'Category', 'SubCat':'subCategory'})
cervical['Filename'] = cervical['Filename'] +'.isyntax'
cervical['train/valid/test']= cervical['AnnotationBatch'].apply(lambda x: 'train' if x <= 35 else 'test') 

cervical_train = cervical.loc[cervical['train/valid/test']=='train']
cervical_test = cervical.loc[cervical['train/valid/test']=='test']

print('Cervical: train ={} , test = {}'.format(len(cervical_train), len(cervical_test)))

```

```python
from sklearn.model_selection import train_test_split
cervical_norm = cervical[(cervical['Category']=='normal_inflammation')]
cervical_mal = cervical[(cervical['Category']=='malignant')]
frames = [cervical_norm, cervical_mal]
cervical_binary = pd.concat(frames)
print(cervical_binary['Category'].value_counts())

cervical_binary_train = cervical_binary[cervical_binary['train/valid/test'] == 'train']
cervical_binary_test = cervical_binary[cervical_binary['train/valid/test'] == 'test']


btrain, bvalid = train_test_split(cervical_binary_train, 
                                  test_size=0.3, random_state=0, 
                                  stratify=cervical_binary_train[['Category','subCategory','StainingSiteID']])
bvalid['train/valid/test'] = 'valid'
df_binary = pd.concat([btrain,bvalid, cervical_binary_test])
df_binary = df_binary.sort_values(by=['Filename'])
cervical_binary = df_binary.reset_index(drop=True)
cervical_binary = cervical_binary[['BiobankID', 'Filename', 'Category', 'train/valid/test']]
cervical_binary['Filename'] = cervical_binary['Filename'].str.replace('.isyntax', '')
cervical_binary = cervical_binary.rename(columns={'BiobankID':'case_id','Filename':'slide_id', 'Category':'label'})
print(cervical_binary['train/valid/test'].value_counts())
cervical_binary.to_csv(root+ '/' + 'weaklysupervised_algo' + '/' + 'cervical_clean_binary_norm_mal.csv', index= False)


cervical_binary = cervical_binary[(cervical_binary['case_id']!= 'IC-CX-00027') & (cervical_binary['case_id']!= 'IC-CX-00063')]

cervical_binary_train = cervical_binary[cervical_binary['train/valid/test'] == 'train']
cervical_binary_test = cervical_binary[cervical_binary['train/valid/test'] == 'test']
cervical_binary_valid = cervical_binary[cervical_binary['train/valid/test'] == 'valid']

print('train:{}\t valid:{} \t test:{}\n\n'.format(len(cervical_binary_train), len(cervical_binary_valid), len(cervical_binary_test)))

train = cervical_binary_train['slide_id']
valid = cervical_binary_valid['slide_id']
test = cervical_binary_test['slide_id']
cervical_binary_split= pd.concat([train.reset_index(drop=True),valid.reset_index(drop=True),test.reset_index(drop=True)], axis = 1, ignore_index=True)
print(cervical_binary_split.head())
cervical_binary_split = cervical_binary_split.rename(columns={cervical_binary_split.columns[0]: 'train',cervical_binary_split.columns[1]: 'val',cervical_binary_split.columns[2]: 'test' })
cervical_binary_split.to_csv(root+ '/' + 'weaklysupervised_algo' + '/' + 'cervical_binary_splits_norm_mal_0.csv')
cervical_binary_split
```

```python
from sklearn.model_selection import train_test_split
train, valid = train_test_split(cervical_train, test_size=0.33, random_state=0, stratify=cervical_train[['Category','subCategory','StainingSiteID']])
valid['train/valid/test'] = 'valid'

df = pd.concat([train,valid, cervical_test])
df = df.sort_values(by=['Filename'])
df = df.reset_index(drop=True)
filenames = ['IC-CX-01256-01.isyntax', 'IC-CX-02287-01.isyntax', 'IC-CX-01964-01.isyntax', 'IC-CX-00176-01.isyntax', 
              'IC-CX-02556-01.isyntax', 'IC-CX-02568-01.isyntax', 'IC-CX-02033-01.isyntax', 'IC-CX-01730-01.isyntax', 
              'IC-CX-01713-01.isyntax', 'IC-CX-02063-01.isyntax', 'IC-CX-03118-01.isyntax', 'IC-CX-02061-01.isyntax',
              'IC-CX-02263-01.isyntax', 'IC-CX-02528-01.isyntax', 'IC-CX-01746-01.isyntax', 'IC-CX-00316-01.isyntax',
              'IC-CX-02065-01.isyntax', 'IC-CX-01860-01.isyntax', 'IC-CX-00033-01.isyntax', 'IC-CX-02931-01.isyntax']

df.loc[df['Filename'].isin (filenames), 'train/valid/test'] = 'valid'
df
```
